## Pivit data

```python
df.groupby(['PROVINCE', 'DIVISION', 'DISTRICT', 'SUB DIVISION'])['ALL SEXES (RURAL)'].mean()
```

## population change

```python
pop_rural_1998 = df['POPULATION 1998 (RURAL)'].sum()
pop_rural_2017 = df['ALL SEXES (RURAL)'].sum()
pop_change = pop_rural_2017 - pop_rural_1998
print('Population is increased', pop_change)
# percent
pop_per = pop_rural_2017 / pop_rural_1998 * 100
print('Population is increased by as of 1998', pop_per, '%')

pop_per = pop_change / pop_rural_1998  * 100
print('Population is increased by %', pop_per)
```
