IQR (Interquartile Range): is the box. Lower quartile is bottom of box. upper quartile is top of the box.
center of an IQR is called median
outliers: outside the 1.5 times above and below is called outlier

Stacked barchart/plot

```python
df.groupby(['pclass', 'sex'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
```

Swarm plot

```python
sns.swarmplot(data=df, x='survived', y='fare', hue='who')
```

Relation of two numaric field

```python
sns.lmplot(data=df, x='fare', y='age')
```

Make groups by age

```python
df['age_group'] = pd.cut(df['age'], [0, 10, 18, 30, 50, 80], labels=['kids', 'child', 'youth', 'middle aged', 'old'])
```

```python
df.head()
```

```python
df.groupby("age_group")['survived'].mean().plot(kind="bar")
```

```python
df.groupby("age_group")['fare'].mean().plot(kind="bar")
```

Pair plot customization

```python
sns.pairplot(df, hue='survived', vars=['age', 'fare', 'pclass'])
```
